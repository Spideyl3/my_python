```python


from transformers import AutoTokenizer, AutoModel
#本地加载大模型
tokenizer = AutoTokenizer.from_pretrained('/content/drive/MyDrive/chatglm3-6b', trust_remote_code=True) #可以自动加载任何tokenizer
model = AutoModel.from_pretrained("/content/drive/MyDrive/chatglm3-6b", trust_remote_code=True, device='cuda') #可以加载任何模型
model = model.eval()
print(model)
```

```python
# import os
# os.system('cat ~/.cache/huggingface/modules/transformers_modules/chatglm3-6b/modeling_chatglm.py')
```

```python
data=input("请输入:")
response, history = model.chat(tokenizer, data, history=[])
print(response)
```
